```python
import pandas as pd
import numpy as np
```

```python
Ecommerce_df = pd.read_csv('data.csv', encoding='latin1')
```

```python
Ecommerce_df.info()
```

```python
Ecommerce_df["Country"].value_counts().head(5)
```

```python
Ecommerce_df["Country"].nunique()
```

```python
Ecommerce_df["Quantity"].describe()
```

```python
import matplotlib.pyplot as plt

plt.boxplot(Ecommerce_df["Quantity"].dropna())
plt.title("Quantity Distribution")
plt.show()
```

```python
Ecommerce_df.fillna({"Description":"Unknown Product"}, inplace=True)
```

```python
Ecommerce_df["UnitPrice"].describe()
```

```python
Ecommerce_df["InvoiceDate"]= pd.to_datetime(Ecommerce_df["InvoiceDate"])
```

```python
Ecommerce_df.info()
```

```python
Ecommerce_df= Ecommerce_df[Ecommerce_df['Quantity']>0]
```

```python
Ecommerce_df= Ecommerce_df[Ecommerce_df['UnitPrice']>0]
```

```python
Ecommerce_df["TotalPrice"] = Ecommerce_df['Quantity'] * Ecommerce_df['UnitPrice']
```

```python
Ecommerce_df.info()
```

```python
Ecommerce_df.dropna(subset=['CustomerID'],inplace=True)
```

```python
Ecommerce_df['CustomerID']=Ecommerce_df['CustomerID'].astype(int)
```

```python
Ecommerce_df.info()
```

```python
Ecommerce_df.describe()
```

```python
snapshot_date = Ecommerce_df['InvoiceDate'].max() + pd.Timedelta(days=1)
customer_data = Ecommerce_df.groupby('CustomerID')['InvoiceDate'].max().reset_index()
customer_data['Recency'] = (snapshot_date - customer_data['InvoiceDate']).dt.days
```

```python
customer_data.head()
```

```python
freq_invoice_per_customer = Ecommerce_df.groupby("CustomerID")['InvoiceNo'].nunique().reset_index()
freq_invoice_per_customer.rename(columns={'InvoiceNo': 'Frequency'}, inplace=True)
```

```python
freq_invoice_per_customer.sort_values(by='Frequency', ascending=False).head(5)
```

```python
freq_invoice_per_customer.shape
```

```python
Monetary_spent = Ecommerce_df.groupby('CustomerID')['TotalPrice'].sum().reset_index()
Monetary_spent.rename(columns={'TotalPrice':'Monetary'}, inplace=True)
```

```python
Monetary_spent.sort_values(by='Monetary', ascending=False).head(5)
```

```python
rfm_df = pd.merge(customer_data,freq_invoice_per_customer,on='CustomerID', how='inner')
```

```python
rfm_df = pd.merge(rfm_df, Monetary_spent, on='CustomerID', how='inner')
```

```python
rfm_df.head()
```

```python
rfm_df.drop('InvoiceDate', axis=1, inplace=True)
```

```python
rfm_df.info()
```

```python
Ecommerce_df.to_parquet('Cleaned_df.parquet',index=False)
```

```python
rfm_df.to_parquet('rfm_df.parquet', index=False)
```
